```python
# Factorial of a number:
n=int(input("Enter the value of n: "))
fact=1
while n>0:
  fact*=n
  n-=1
print("Factorial =",fact)
```

```python
# Write a python program for a fibbonacci serise using while loop upto n terms
term=int(input("Enter the value of n: "))
i,a,b=0,0,1
while i<=term:
  print(a,end=" ")
  c=a+b
  a=b
  b=c
  i+=1
```

```python
# Reverse a number
n,revnum=int(input("Enter the value of n: ")),0
while n>0:
  revnum*=10
  revnum+=(n%10)
  n//=10
print("Number Reversed =",revnum)
```

```python
# Check Palindrome number:
def reverse(n):
  revnum=0
  while n>0:
    revnum*=10
    revnum+=(n%10)
    n//=10
  return revnum
n=int(input("Enter the value of n: "))
if n==reverse(n):
  print("Number is Palindrome")
else:
  print("Number is Not a Palindrome")

```

```python
st,revst=input("Enter a String: "),""
i=len(st)
while i>0:
  revst=revst+st[(i-1)]
  i-=1
if(st==revst):
  print("String is Palindrome")
else:
  print("String is Not a Palindrome")
```

```python
# Count Number of digits
num=int(input("Enter a number: "))
count=0
while num>0:
  count+=1
  num//=10
print("The number Has",count,"Digits.")
```

```python
# Search an element in list using while loop
nums=[10,20,30,40,50]
key=int(input("Enter the value of key: "))
i=0
while i<5:
  if nums[i]==key:
    print("Element found at index",i)
    break
  i+=1
else:
  print("Element not found")
```
